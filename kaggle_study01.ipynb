# 결측치 처리에 대해 우선 임시로 dropna함.

```
train_data = train_data.drop(['inst_id','sido','sgg'],axis=1)
test_data = test_data.drop(['inst_id','sido','OC','sgg'],axis=1)
```

```
train_data = train_data.fillna(0)
test_data = test_data.fillna(0) 
```

```
print(train_data.shape)
print(train_data.dropna().shape)
print(test_data.shape)
print(test_data.dropna().shape)
```

```
y_train = train_data[['OC']]
y_train = y_train.replace(['open', ' close'], [1,0]).astype('int')
print(y_train.shape)
train_data = train_data.drop(['OC'],axis=1)
x_train = train_data
x_test = test_data
```

# 데이터의 차원을 줄이기 위해 각 변수별 상관관계 분석

```
x_train.corr()
```

# 변수들간의 상관관계가 매우 적은것들을 drop해주자.

```
x_train = x_train.drop(['openDate','profit1','receivableL1'],axis=1)
x_test = x_test.drop(['openDate','profit1','receivableL1'],axis=1)
```

```
print(train_data['instkind'].unique())
print(train_data['ownerChange'].unique())
```

```
x_train['instkind']
```

```
x_train['ownerChange'] = x_train['ownerChange'].replace(['same', 'change'], [1,0]).astype('int')
x_test['ownerChange'] = x_test['ownerChange'].replace(['same', 'change'], [1,0]).astype('int')
x_train['instkind'] = x_train['instkind'].replace(['nursing_hospital','general_hospital','traditional_clinic','clinic','traditional_hospital','dental_clinic','hospital'], ['0','1','2','3','4','5','6'])
x_test['instkind'] = x_test['instkind'].replace(['nursing_hospital','general_hospital','traditional_clinic','clinic','traditional_hospital','dental_clinic','hospital'], ['0','1','2','3','4','5','6'])
```

```
#x_test['ownerChange'] = x_test['ownerChange'].astype(float)
x_test = x_test.replace(['1,637', '1,816','1,065','1,547','1,104'], [1637,1816,1065,1547,1104]).astype(float)
```

```
x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train)
```

```
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from sklearn.svm import SVC

log_clf = LogisticRegression()
rnd_clf = RandomForestClassifier(n_estimators=10)
svm_clf = SVC(probability=True)

eclf1 = VotingClassifier(estimators=[('lr',log_clf),('rf',rnd_clf),('svc',svm_clf)], voting='soft')
```

```
eclf1.fit(x_train, y_train)
y_pred = eclf1.predict(x_test)
```

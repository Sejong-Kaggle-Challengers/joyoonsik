# 결측치 처리에 대해 우선 임시로 dropna함.

```
train_data = train_data.drop(['inst_id','sido','sgg'],axis=1)
test_data = test_data.drop(['inst_id','sido','OC','sgg'],axis=1)
```

```
train_data = train_data.dropna()
test_data = test_data.dropna()
```

```
print(train_data.shape)
print(train_data.dropna().shape)
print(test_data.shape)
print(test_data.dropna().shape)
```

```
y_train = train_data[['OC']]
train_data = train_data.drop(['OC'],axis=1)
x_train = train_data
x_test = test_data
```

# 변수들간의 상관관계가 매우 적은것들을 drop해주자.

```
x_train = x_train.drop(['openDate','profit1','receivableL1','instkind'],axis=1)
x_test = x_test.drop(['openDate','profit1','receivableL1','instkind'],axis=1)
```

```
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
```

```
#One hot encoding 필요한 Feature
print(train_data['instkind'].unique())
print(train_data['ownerChange'].unique())
print(y_train['OC'].unique())
```

```
y_train = y_train.reset_index(drop=True)
for i in range(y_train.shape[0]):
  if y_train['OC'][i] == "open":
    y_train['OC'][i] = 1
  else:
    y_train['OC'][i] = 0

x_train = x_train.reset_index(drop=True)
for i in range(x_train.shape[0]):
  if x_train['ownerChange'][i] == "same":
    x_train['ownerChange'][i] = 1
  else:
    x_train['ownerChange'][i] = 0

x_test = x_test.reset_index(drop=True)
for i in range(x_test.shape[0]):
  if x_test['ownerChange'][i] == "same":
    x_test['ownerChange'][i] = 1
  else:
    x_test['ownerChange'][i] = 0
```
